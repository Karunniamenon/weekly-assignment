### 4.Create a histogram of "Age"

```python
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(8,8))
plt.hist(data['Age'],color='red',rwidth=.9)
plt.title('Age of patients',fontsize=18)
plt.xlabel('Age')
plt.ylabel('Number')
plt.xticks()
plt.yticks()
```

### INFERENCE:-
     More than 90 percent belongs to age category between 21-50.
     Old age frequency distribution is very low compared with other categories.

### 5.Create a scatter plot of "BMI" vs "Glucose"

```python
plt.figure(figsize=(8,8))
plt.scatter(data['BMI'],data['Glucose'],s=10)
plt.title('Plot of BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
```

### INFERENCE:-
     BMI distribution is clusterred in the range of 20-50.
     Glucose level mainly ranged between 50-200.
